# **Introduction:**

**What is Exploratory Data Analysis ?**

Exploratory Data Analysis or (EDA) is understanding the data sets by summarizing their main characteristics often plotting them visually. This step is very important especially when we arrive at modeling the data in order to apply Machine learning. Plotting in EDA consists of Histograms, Box plot, Scatter plot and many more. It often takes much time to explore the data. Through the process of EDA, we can ask to define the problem statement or definition on our data set which is very important.

# What are our objectives ?

1. Get maximum insights from a data set
2. Uncover underlying structure
3. Extract important variables from the dataset
4. Detect outliers and anomalies(if any)
5. Test underlying assumptions
6. Determine the optimal factor setting

# Insights
Insights we found

`academic progress vs attendance`

```python

# Assuming your CSV file has columns 'Gender' and 'Attendance'
# Create a pivot table to summarize the data for the stacked bar chart
pivot_table = df.pivot_table(index='Gender', columns='What is your current percentage of attendance in Knowledge Pro ?', aggfunc='size', fill_value=0)

# Create the stacked bar chart
plt.figure(figsize=(8, 6))  # Adjust the figure size as needed
sns.set(style="whitegrid")
sns.set_palette("Set2")

# Plot the stacked bar chart
pivot_table.plot(kind='bar', stacked=True)

# Customize the plot
plt.title('Gender vs. Attendance')
plt.xlabel('Gender')
plt.ylabel('Percentage')
plt.legend(title='Attendance', bbox_to_anchor=(1, 1))

plt.show()
```

```python
# Assuming your CSV file has a 'Department' column
# Count the number of attendees in each department
department_counts = df['Which course do you study ?'].value_counts()

# Find the most attending department
most_attending_department = department_counts.idxmax()
most_attending_count = department_counts.max()

# Create a bar chart
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.bar(department_counts.index, department_counts.values)

# Customize the plot
plt.title('Most Attending Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.text(most_attending_department, most_attending_count, f'Most Attending: {most_attending_count}', ha='center', va='bottom')

plt.show()
```

```python
df_sorted = df.sort_values(by='How would you rate your academic progress in the past semester?', ascending=True)

# Create a bar chart to visualize the effect on mental health for each department
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.bar(df_sorted['Which course do you study ?'], df_sorted['How would you rate your academic progress in the past semester?'], color='red')

# Customize the plot
plt.title('Most change in academically effected Department')
plt.xlabel('Department')
plt.ylabel('Academical progress')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

plt.show()
```
